```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
from matplotlib import style
```

# Bar graph

```python
%matplotlib inline
x1=np.array([5,5,10])
y1=np.array([12,6,6])
x2=np.array([7,3,10])
y2=np.array([21,3,7])
plt.bar(x1,y1,label="bar one",color='m')
plt.bar(x2,y2,label="bar two",color='c')
plt.legend()
plt.xlabel("bar number")
plt.ylabel("bar values")
plt.title("sample bar graph")
plt.show()
```

# scatter plot

```python
%matplotlib inline
x1=np.array([2,3,10])
y1=np.array([12,1,4])
plt.scatter(x1,y1,label="scat",color='r')
plt.xlabel("x scat")
plt.ylabel("y scat")
plt.title("scatterplot")
plt.show()
```

# area plot(pie chart)

```python
%matplotlib inline
a=[0,5,1,2]
b=['green','blue','black','red']
colors=['g','b','k','r']
plt.pie(a,labels=b,colors=colors,startangle=90,shadow=True)
plt.title("pie chart")
plt.show()
```

# Linear graph

```python
%matplotlib inline
x=np.array([5,1,3])
y=np.array([3,7,1])
plt.plot(x,y,label="line")
plt.title("sample linear graph")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend()
plt.show()
```
